پیاده سازی صف به کمک آرایه
</font>
<p></p>
برای پیاده سازی صف به کمک آرایه ابتدا یک آرایه در نظر می گیریم و دو متغیر که یکی ابتدای صف 
($first$)
و دیگری تعداد عناصری که در صف قرار دارند
($num$)
را نشان می‌دهند.
<br>
هنگامی که عنصر جدیدی اضافه می شود آن را در انتهای صف 
 $num$ قرار می دهیم و سپس مقدار 
را افزایش می دهیم. برای حذف کردن عنصر ابتدایی صف نیز مقدار $num$ را یکی کم می کنیم.
<p></p>
$enQ(x)$: درج عنصر x در انتهای صف
<br>
$deQ()$: حذف و بازگرداندن اولین عنصر صف .
<br>
$front()$: اولین عنصر صف  را باز می گرداند.
<br>
$size()$: تعداد عناصر موجود در صف  را باز می گرداند.
<br> 
$is\_empty()$: خالی بودن یا نبودن صف را مشخص می کند.
<br>
$is\_full()$: پر بودن یا نبودن صف را مشخص می کند.
<p></p>
تمامی اعمال در $O(1)$ 
انجام می‌شوند

```python
class Queue:
    def __init__(self, size):
        self.size = size
        self.Q = [0] * size
        self.num = 0
        self.first = 0
    
    def enQ(self, item):
        if self.num >= self.size:
            raise Exception("Queue overflow")
        self.Q[(self.num + self.first) % self.size] = item
        self.num += 1
        print self.Q, self.first, self.num # Comment this line

        
    def deQ(self):
        if self.num == 0:
            return None
        item = self.Q[self.first]
        self.first = (self.first + 1) % self.size
        self.num -= 1
        print self.Q, self.first, self.num # Comment this line
        return item
    
    def front(self):
        if self.num == 0:
            return None
        return self.Q[self.first]
    
    def is_empty(self):
        return self.num == 0
    
    def size(self):
        return self.num
    
    def is_full(self):
        return self.num >= self.size
```

پیاده سازی صف دو طرفه به کمک آرایه
</font>
<p></p>
برای پیاده سازی صف دو طرفه به کمک آرایه ابتدا یک آرایه در نظر می گیریم و دو متغیر که ابتدا و تعداد عناصر موجود در صف را نشان می دهند.
<p></p>
$push\_front(x)$: درج عنصر $x$ در جلوی صف
<br>
$push\_back(x)$: درج عنصر $x$ در عقب صف
<br>
$pop\_front(x)$: حذف و بازگرداندن عنصر جلوی صف 
<br>
$pop\_back(x)$: حذف و بازگرداندن عنصر عقب  صف 
<br>
$size()$: تعداد عناصر موجود در صف  را باز می گرداند.
<br> 
$is\_empty()$: خالی بودن یا نبودن صف را مشخص می کند.
<br>
$is\_full()$: پر بودن یا نبودن صف را مشخص می کند.
<p></p>
تمامی اعمال در $O(1)$ انجام می‌شوند

```python
class TwoWayQueue:
    def __init__(self, size):
        self.size = size
        self.Q = [0] * size
        self.num = 0
        self.first = 0
    
    def push_back(self, item): # Like one-way queue
        if self.num >= self.size:
            raise Exception("Queue overflow")
        self.Q[(self.num + self.first) % self.size] = item
        self.num += 1
        print self.Q, self.first, self.num # Comment this line

    def push_front(self, item): # New
        if self.num >= self.size:
            raise Exception("Queue overflow")
        self.first = (self.first - 1) % self.size
        self.Q[self.first] = item
        self.num += 1
        print self.Q, self.first, self.num # Comment this line

    def pop_front(self): # Like one-way queue
        if self.num == 0:
            return None
        item = self.Q[self.first]
        self.first = (self.first + 1) % self.size
        self.num -= 1
        print self.Q, self.first, self.num # Comment this line
        return item
        
    def pop_back(self): # New
        if self.num == 0:
            return None
        self.num -= 1
        print self.Q, self.first, self.num # Comment this line
        return self.Q[(self.num + self.first) % self.size]
    
    def is_empty(self):
        return self.num == 0
    
    def size(self):
        return self.num
    
    def is_full(self):
        return self.num >= self.size
```

پشته
<p></p>
</font>
دنباله ای از عناصر که ترتیب در آن ها مهم است. تنها به انتهای پشته می توان عنصر اضافه کرد  و تنها به انتهای آن دسترسی داریم.به عنوان مثال ظرف های کثیفی را در نظر بگیرید 
که بر روی هم قرار دارند  و قصد شستن آنان را دارید  در هر مرحله اگر بخواهید ظرفی را بردارید بالاترین آنان را انتخاب می کنید و آن را می شویید و اگر ظرف کثیفی را بخواهید 
به آنان اضافه کنید در بالای تمام آنان قرار می دهید.
<br>
$push(x)$:درج عنصر $x$ در انتهای پشته
<br>
$pop()$: حذف و بازگرداندن آخرین عنصر پشته
<br>
$top()$: آخرین عنصر پشته را باز می گرداند.
<br>
$size()$: تعداد عناصر موجود در پشته را باز می گرداند.	
$is\_empty()$: خالی بودن یا نبودن پشته را مشخص می کند.
<br>
$is\_full()$: پر بودن یا نبودن پشته را مشخص می کند.
<p></p>
تمام اعمال در $O(1)$ انجام می شوند.
پیاده سازی پشته به کمک آرایه
برای پیاده سازی پشته یک آرایه در نظر می گیریم و از یک متغیر مانند $num$ که تعداد عناصر موجود در پشته را نشان می دهد برای اضافه کردن و حذف عناصر انتهای پشته کمک می گیریم.
</font></div>

```python
class Stack:
    def __init__(self, size):
        self.size = size        # Size of stack
        self.S = [0] * size     # Stack array
        self.num = 0            # Number of elements in Stack
    
    def push(self, item):
        if self.num >= self.size:
            raise Exception("Stack overflow")
        self.S[self.num] = item
        self.num += 1
        
    def pop(self):
        if self.num == 0:
            raise Exception("Stack empty")
        self.num -= 1
        return self.S[self.num]
    
    def top(self):
        if self.num == 0:
            raise Exception("Stack empty")
        self.num -= 1
        return self.S[self.num]

    
    def size(self):
        return self.num
    
    def is_full(self):
        return self.num >= self.size
    
    def is_empty(self):
        return self.num == 0
```

لیست پیوندی
<p></p>
</font>
دنباله ای از عناصر که ترتیب در آن ها مهم است. درج و حذف در تمام لیست ممکن است و در $O(n)$ انجام می شود. درلیست پیوندی هر عنصر به عضو بعدی خود اشاره می کند.
<br>
در لیست پیوندی دو سویه علاوه بر عضو بعدی هر عنصر به عضو قبلی نیز اشاره می کند. 
<br>
در لیست پیوندی حلقه ای عضو آخر به عضو اول به عنوان عضو بعدی و عضو اول به عضو آخر به عنوان عضو قبلی اشاره می کند.
<br>
$insert\_after(data,x):$ درج عنصر $data$ بعد از عنصر $x$
<br>
$delete(x):$ حذف و بازگرداندن عنصر $x$
<br>
$find(val):$ اولین عنصر با $data$ مساوی با $val$ را باز می گرداند
<br>
$get(ind):$ عنصر $ind$ام لیست را باز می گرداند.
<br>
$size():$ تعداد عناصر موجود در  لیست را باز می گرداند.	
$is\_empty():$ خالی بودن یا نبودن لیست را مشخص می کند
<p></p>
<font size=6>
پیاده سازی لیست به کمک کلاس
</font>
<br>
یک کلاس برای $node$ های لیستمان می گیریم و در به ازای هر $node$ اطلاعات لازم و $node$ بعد و قبل آن را به ترتیب در $next$ و $prev$ آن قرار می دهیم.
حال به کمک کلاس $list$ که در آن $node start$ به عنوان $node$ قبل از اولین عنصر و بعد از آخرین عنصر قرار دارد لیست پیوندی دوسویه را پیاده سازی می کنیم.
</font></div>


```python
class Node:
  def __init__(self, data):
    self.data = data  
    self.next = None  
    self.prev = None  

  def get_data(self):
    return self.data

class List:
    def __init__(self):        
        self.start = Node(None)
        self.start.next = self.start
        self.start.prev = self.start
        self.n = 0
        
    def get(self,ind):
        if ind >=  self.size() : 
          raise Exception('Out of list')
        x = self.start.next
        for i in range(ind) :
          x=x.next
        return x
    
    def insert_after(self, x, data):
        y = Node(data)
        self.n += 1
        y.prev = x
        y.next = x.next
        x.next = y
        y.next.prev = y
        return y
        
    def delete(self, x):
        self.n -= 1
        x.prev.next = x.next
        x.next.prev = x.prev
        return x
        
    def find(self, val):
        x = self.start.next
        for i in range(self.size()) :
          if x.data == val :
            return x
          x=x.next
        return None
    
    def size(self):
      return self.n
      
    def is_empty(self):
      return self.n==0
```
