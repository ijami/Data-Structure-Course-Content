چرا مرتب سازی سریع تصادفی؟
</font>
<br>
همان طور که جلسه قبل دید، حالت هایی برای مرتب‌سازی سریع وجود دارد که باعث می‌شود عملکرد آن قابل‌قبول نباشد.
برای اصلاح این مشکل از تکنیکی به نام <b> نمونه‌برداری تصادفی </b> استفاده می‌کنیم.
بدین معنی که به جای این که همواره عنصر محور را آخرین عنصر آرایه در نظر بگیریم، به صورت تصادفی یکی از عناصر آرایه را به عنوان محور انتخاب می‌کنیم. سپس همان رونده گذشته را تکرار می کنیم.
<br>
نکته مهمی که در این قسمت وجود دارد، تحیلیل بهینه بودن این روش  است که در ادامه می آید: (تسلط به مطالب جزوه قبل به فهم بهتر مطالب این قسمت کمک خواهد کرد)

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>

<font color=#FF7500 size=6>
<p></p>
پیش نیاز تحلیل: احتمال</font>
<hr>
برای این که بتوانیم تحیلیل جامعی از این الگوریتم ارائه دهیم، ابتدا شما را ا چند مفهوم در آمار و احتمال آشنا می‌کنیم.
<ul>
<li> <b> فصای نمونه</b>: تمامی خروجی های ممکن یک رویداد، که آن را با $\Omega$ نشان می دهند و دو خاصیت زیر را دارد:
    <ol>
    <li> احتمال وقوع هر خروجی بزرگتر مساوی صفر است
    $$\exists i \in \Omega : p(i) >= 0$$
    <li> مجموع احتمال خروجی ها برابر یک است
    $$\sum_{i \in \Omega} p(i) = 1$$
    </ol>
    به عنوان مثال فضای نمونه اندیس عنصری که به عنوان محور در یک آرایه n تایی انتخاب می شود به صورت زیر است:
    $$\Omega = \{ 0,1,....n-1 \}$$
<li> <b>پیشامد</b>: هر زیرمجموعه ای از فضای نمونه را یک پیشامد می نامند. احتمال پیشامد $S \in \Omega$ برابر است با
$$\sum_{i \in S} p(i) $$
به عنوان مثال احتمال این پیشامد که محور به گونه‌ای انتخاب شود که اندازه بخش کوچکتر حداقل $1/4 $ آرایه کلی باشد برابر با $1/2$  است.
<li> <b>متغییر تصادفی</b>: یک تابع  $X : \Omega \to \mathbb{R}$ است
<br>
به عنوان مثال اندازه بخش هایی که در نتیجه افراز مرتب‌سازی سریع به دست می آید.
<li> <b>امید ریاضی</b>: به نوعی میانیگن وزن دار است روی فضای نمونه است، تعریف دقیق آن به صورت زیر است:
    $$E[X] = \sum_{i \in \Omega} p(i).X(i)$$
به عنوان مثال امید ریاضی طول بخش یکی از بخش ها که در اولین گام مرتب سازی سریع بر روی یک آرایه n عضوی اجرا می‌شود برابر با $n/2$ است.
<li> <b>خطی بودن امید ریاضی</b>: ادعا می کند که اگر $X_1$ تا $X_n$ متغییر تصادفی هایی باشند که روی $\Omega$ تعریف می شوند، آنگاه خواهیم داشت:
$$E[ \sum_{i=1}^{n} X_i] = \sum_{i=1}^{n} E[X_i]  $$
<li> <b> تابع مشخصه </b> : یک نوع خاص از متغییر تصادفی که به صورت زیر تعریف می شود:
$$X : \Omega \to {0,1}$$
به عنوان مثال مقایسه شدن یا نشدن دو عنصر در زوند مرتب‌سازی سریع
</ul>

<p></p>
مقدمات: تعریف مسئله به صورت احتمالاتی
</font>
<p></p>
در هر گام مرتب سازی سریع، محور انتخابی نقش کلیدی را بازی می کند، چون می توان با داشتن آن دو بخش حاصله از افراز را یافت. بنابراین می توان گفت که با داشتن دنباله ای از کلید ها می توان کل روند مرتب سازی سریع سریع را باز سازی کرد. بنابراین:
<br>
<b> فضای نمونه $\Omega$</b>  تمامی دنباله های که از انتخاب محور ها حاصل می شود.
<br>
<br>
همچنین در روند مرتب‌سازی سریع، فرایندی که زمان بر است، مقایسه کلید ها با هم دیگر در بخش افراز است، بنابراین برای محاسبه زمان اجرای مرتب سازی‌سریع می توان متغییر تصادفی زیر را در نظر گرفت:
<br>
<b> متغییر تصادفی $C(\sigma)$</b> برای هر $\sigma \in \Omega$ برابر است با تعداد مقایسه ها هنگامی که دنباله محور ها $\sigma$ باشد
<br>
<br>
پس اگر بخواهیم زمان اجرای مرتب سازی سریع را محاسبه کنیم، می توانیم مقدار $C(\sigma)$ را در حالت میانگین حساب کنیم، به عبارت دیگر $E[C(\sigma)]$ را به دست بیاوریم.
<br>
 اما این کار به صورت مستقیم بسیار سخت و پیچیده است

روش طلایی: شکستن مسئله
</font>
<p></p>
هنگامی که با یک مسئله بسیار سخت روبه‌رو می شویم، می توانیم آن را به قطعات کوچکتر تقسیم کنیم، آن ها را حل کنیم و سپس با استفاده از آن ها به حل مسئله نهایی برسیم. (از این روش در درس طراحی الگورتیم ها نیز استفاده خواهید کرد.)
<br>
در این مسئله قطعات کوچکتر ما تابع مشخصه است که به صورت زیر تعریف می شود:
<br>
$X_{ij} (\sigma) $= تعداد دفعاتی که $z_i$ام و $z_j$ با هم مقایسه می شوند.
<br>
($z_i$ ، $i$امین عنصر در آرایه مرتب شده است)
<br>
مقدار $X_{ij} (\sigma) $ صفر یا یک است، زیرا دو عنصر فقط هنگامی با هم مقایسه می شوند که یکی از آن ها محور باشد، و پس از این مقایسه عنصر دیگر در بخش جداگانه ای از عنصر محور قرار می گیرد ، بنابراین هیچگاه دوباره با آن مقایسه نخواهد شد.
<br>
طبق تعریفی که برای $C(\sigma)$ ارائه شد، می توان تعداد مقایسه ها را این گونه بیان کرد که چند عنصر داریم که $X_{ij} (\sigma) $ آن ها یک است، یا به عبارت دیگر:
    $$ C(\sigma) =  \sum_{i=1}^{n-1} \sum_{j=i+1}^{n} X_{ij}(\sigma)$$
حال با توجه به <b>خطی بودن امید ریاضی</b> داریم:
 $$ E[C(\sigma)] =  E[\sum_{i=1}^{n-1} \sum_{j=i+1}^{n} X_{ij}(\sigma)] = \sum_{i=1}^{n-1} \sum_{j=i+1}^{n} E[X_{ij}(\sigma)]$$
 که نشان می دهیم می توان  $E[X_{ij}(\sigma)]$ را به راحتی به دست آورد

<p></p>
محاسبات ریاضی
</font>
<hr>
از آنچایی گه طبق تعریف داریم:
$$ E[X_{ij}] = p_{ij}.1 + (1-p_{ij}).0$$
بنابراین کافی است $p_{ij}$ یعنی احتمال این که  $z_i$ , $z_j$ با هم مقایسه شوند را به دست آورد.
<br>
هنگامی یکی از $z_i$ یا $z_j$ به عنوان محور انتخاب می شوند، که هیچ یک از عناصر بینشان به عنوان محور انتخاب نشده باشد، بنابراین خواهیم داشت:
$$p_{ij} = \frac{2}{j-i+1}$$
<br>
بنابراین داریم:
 $$ E[C(\sigma)] = \sum_{i=1}^{n-1} \sum_{j=i+1}^{n} \frac{2}{j-i+1} = 2\sum_{i=1}^{n-1} \sum_{j=i+1}^{n} \frac{1}{j-i+1}$$
 با استفاده از تغییر متغییر $k=j-i$ داریم
  $$= 2\sum_{i=1}^{n-1} \sum_{k=1}^{n-i} \frac{1}{k+1} \leq   2\sum_{i=1}^{n-1} \sum_{k=2}^{n} \frac{1}{k} \leq 2n\sum_{k=2}^{n} \frac{1}{k}$$
که $\sum_{k=1}^{n} \frac{1}{k}$ به سری هارمونیک مشهور است و با استفاده از تقریب انتگرال ثابت می شود:
$$\sum_{k=1}^{n} \frac{1}{k} \leq ln(n)+1$$
در نتیجه خواهیم داشت:
$$E[C(\sigma)] \leq 2n\sum_{k=2}^{n} \frac{1}{k}\leq 2.n.ln(n)$$
بنابراین می‌توان نتیجه گرفت که مرتب سازی سریع از $\theta(nlogn)$ است.
