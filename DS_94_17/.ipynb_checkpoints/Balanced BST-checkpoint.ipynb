<p></p>

<hr>
یک درخت سیاه-قرمز
(یا Red-Black Tree)
یک د.د.ج است که خواص زیر را داشته باشد:
<ul>
    <li> هر راس یک رنگ دارد که یا سیاه است یا قرمز. </li>
    <li> راس ریشه همیشه سیاه است. </li>
    <li> هر راس غیر
    NIL
    دقیقا ۲ فرزند دارد (یعنی تمامی برگ ها 
    NIL
   یا پوچ هستند)
     و رنگ تمامی برگ‌ها حتما سیاه است.
     </li>
    <li> اگر راسی قرمز باشد، حتما هر دو فرزند آن سیاه هستند. </li>
    <li> برای هر راس، هر مسیر ساده از آن به یکی از یرگ‌های زیردرختش دارای تعداد مساوی از راس‌های سیاه است. </li> 
</ul>
<img src="images/1.jpg" style="float:center;">
این خواص باعث می‌شوند طول بلندترین مسیر از ریشه به یکی از برگ‌ها حداکثر دو برابر طول کوتاه‌ترین مسیر از ریشه به یکی از برگ‌ها باشد که توازن خوبی را در درخت ایجاد می‌کند.
<br>
(تعداد رئوس سیاه مسیرها که مساوی است، از آنجایی که فرزندان رئوس قرمز حتما سیاه هستند، تعداد رئوس قرمز حداقل ۰ و حداکثر به تعداد رئوس سیاه است که خاصیت بالا را نتیجه می‌دهد)

چرخش
</font>
<p></p>

<hr>
برای نگه داشتن خواص درخت قرمز-سیاه در زمان انجام عملیات درج و حذف، یک عمل به اسم «چرخش» به راست و چپ تعریف می‌شود که به این صورت است:
<br>
در چرخش به راست روی 
$a$
فرض می‌کنیم فرزند چپ آن 
(مثلا $b$)
پوچ
(NIL)
نیست، اگر فرزند سمت راست 
$a$
را 
$a_r$ 
بنامیم، و فرزند چپ 
$b$
را 
$b_l$ 
و فرزند راست 
$b$
را 
$x$
بنامیم، با چرخش به راست:
<ol>
    <li> 
    $b$
    می‌رود جای
    $a$. 
    </li>
    <li> 
    $a$
    فرزند راست 
    $b$
    می‌شود.
    </li>
    <li>
    $x$
    (و زیردرختش) فرزند چپ 
    $a$
    می‌شود.
    </li>
    <li>
    $a_r$
    (و زیردرختش) فرزند راست 
    $a$
    باقی می‌ماند.
    </li>
    <li>
    $b_l$
    (و زیردرختش) فرزند چپ 
    $b$
    باقی می‌ماند.
    </li>
</ol>
چرخش به چپ هم دقیقا برعکس همین است.
     <img src="images/1.gif" style="float:center;">

```python
# look at this code only as a psuedo code


def _left_rotate(self, x):
    y = x.right
    x.right = y.left  # turn y's left subtree into x's right subtree

    if y.left != self.nil:
        y.left.p = x
    y.p = x.p  # link x's parent to y

    if x.p == self.nil:
        self._root = y
    elif x == x.p.left:
        x.p.left = y
    else:
        x.p.right = y

    y.left = x  # put x on y's left
    x.p = y
```

درج
</font>
<p></p>

<hr>
برای درج یک عنصر
$x$
در یک درخت قرمز-سیاه، اول مانند یک د.د.ج عادی آن را درج می‌کنیم و رنگش را قرمز میکنیم و ۲ فرزند سیاه پوج 
(NIL)
برای آن در نظر میگیریم.
<br>
تا اینجا تمام خواص گفته شده برای درخت قرمز-سیاه رعایت شده جز چهارمی: ممکن است راس پدر
$x$
(که آن را
$p[x]$
می‌نامیم)، قرمز رنگ باشد.
در این صورت باید با تغییراتی در رنگ‌ها و با چرخش‌ها این مشکل را حل کنیم.
<br>
چون بقیه‌ی درخت قبلا درست بوده پدر پدر
$x$
(یعنی 
$p[p[x]]$)
قظعا سیاه است.
<br>
اگر فرض می‌کنیم 
$p[x]$
فرزند چپ
$p[p[x]]$
است
چند حالت پیش می‌آید:
<br>
<ol>
    <li>
        اگر فرزند راست 
        $p[p[x]]$
        ($y$)
        هم قرمز باشد، 
        $y$ و $p[x]$
        را سیاه می‌کنیم و 
        $p[p[x]]$
        را قرمز میکنیم، حالا فرض می‌کنیم 
        $p[p[x]]$
       عنصر درج شده است و همین عملیات را برای آن اجرا می‌کنیم.
       <img src="images/2.gif" style="float:center;">
       (اگر به ریشه رسیدیم ریشه را سیاه می‌کنیم و عملیات به پایان می‌رسد).
    </li>
    <li>
        اگر
        $y$
        سیاه باشد
        <ul>
            <li>
                چک می‌کنیم که 
                $x$ 
                فرزند چپ 
                $p[x]$
                هست یا نه، اگر بود که هیچ، اگر نبود روی 
                $p[x]$
                چرخش به چپ انجام می‌دهیم تا 
                $p[x]$
                فرزند چپ 
                $x$
                شود. و از این پس فرض می‌کنیم 
                $p[x]$
                همان 
                $x$
                است.
            </li>
        </ul>
        $p[x]$
        را سیاه می‌کنیم،
        $p[p[x]]$
        را قرمز می‌کنیم و در نهایت روی 
        $p[p[x]]$
        یک چرخش به راست انجام می‌دهیم و عملیات به پایان می‌رسد.
        <img src="images/3.gif" style="float:center;">
    </li>
</ol>     
در صورتی که
$p[x]$
فرزند راست پدرش باشد همه‌ی این کارها به صورت قرینه انجام می‌شود.

```python
class RBNode(object):
    """
    Node in a red black Tree
    """

    def __init__(self, key):
        self.key = key
        self.red = False
        self.left = None
        self.right = None
        self.p = None  # parent


class RedBlackTree(object):
    """
    A red black tree. See Cormen, Leiserson, Rivest, Stein 3rd edition chapter 13.
    """
    def __init__(self, r_b_node=RBNode):
        self.r_b_node = r_b_node

        self._nil = self._create_node(None)  # nil is used for all of our leaves
        self._root = self.nil

    @property
    def root(self):
        return self._root

    @property
    def nil(self):
        return self._nil

    def _create_node(self, key):
        return self.r_b_node(key)

    def _left_rotate(self, x):
        y = x.right
        x.right = y.left  # turn y's left subtree into x's right subtree

        if y.left != self.nil:
            y.left.p = x
        y.p = x.p  # link x's parent to y

        if x.p == self.nil:
            self._root = y
        elif x == x.p.left:
            x.p.left = y
        else:
            x.p.right = y

        y.left = x  # put x on y's left
        x.p = y

    def _right_rotate(self, y):
        # note that this is symmetric to left rotate
        x = y.left
        y.left = x.right

        if x.right != self.nil:
            x.right.p = y
        x.p = y.p

        if y.p == self.nil:
            self._root = x
        elif y == y.p.right:
            y.p.right = x
        else:
            y.p.left = x

        x.right = y
        y.p = x

    def insert_key(self, key):
        self._insert_node(self._create_node(key))

    def _insert_node(self, z):
        y = self.nil
        x = self.root

        while x != self.nil:
            y = x
            if z.key < x.key:
                x = x.left
            else:
                x = x.right

        z.p = y
        if y == self.nil:
            self._root = z
        elif z.key < y.key:
            y.left = z
        else:
            y.right = z

        z.left = self.nil
        z.right = self.nil
        z.red = True
        self._insert_fix_up(z)

    def _insert_fix_up(self, z):
        while z.p.red:
            if z.p == z.p.p.left:
                y = z.p.p.right
                if y.red:
                    z.p.red = False
                    y.red = False
                    z.p.p.red = True
                    z = z.p.p
                else:
                    if z == z.p.right:
                        z = z.p
                        self._left_rotate(z)
                    z.p.red = False
                    z.p.p.red = True
                    self._right_rotate(z.p.p)
            else:
                y = z.p.p.left
                if y.red:
                    z.p.red = False
                    y.red = False
                    z.p.p.red = True
                    z = z.p.p
                else:
                    if z == z.p.left:
                        z = z.p
                        self._right_rotate(z)
                    z.p.red = False
                    z.p.p.red = True
                    self._left_rotate(z.p.p)

    def search(self, key, x=None):
        """
        searches for the key in the subtree of x
        :param key: the key which we are searching for
        :param x: searching in the subtree rooted at x
        :return: self.nil if it cannot find it
        """
        if x is None:
            x = self.root

        while x != self.nil and key != x.key:
            if key < x.key:
                x = x.left
            else:
                x = x.right

        return x

    def minimum(self, x=None):
        if x is None:
            x = self.root

        while x.left != self.nil:
            x = x.left
        return x

    def maximum(self, x=None):
        if x is None:
            x = self.root

        while x.right != self.nil:
            x = x.right
        return x
```

```python

def check_correct_properties(self):
    def is_correct_red_black_node(node):
        """
         returns black number of the subtree rooted at node
        """

        if (node.left is None) != (node.right is None):
            raise Exception("All nodes must have both left and right child or none of them")

        if (node.left is None) and (node.right is None) and node.red:
            raise Exception("Red nodes must have children")

        if node.red and node.left and node.right:
            if node.left.red or node.right.red:
                raise Exception("Children of red nodes must be black")

        if node.left and node.right:
            if (self.nil != node.left and node != node.left.p) or \
                    (self.nil != node.right and node != node.right.p):
                raise Exception("Parent of node is not correctly set")

            left_counts = is_correct_red_black_node(node.left)
            right_counts = is_correct_red_black_node(node.right)

            if left_counts != right_counts:
                raise Exception("Black number of children are not equal")

            return left_counts
        else:
            return 0

    try:
        is_correct_red_black_node(self.root)
        if self.root.red is True:
            raise Exception("Root of the tree must be black")
    except Exception, e:
        print "Red black tree is not OK because: %s" % e.message
        return False

    return True

RedBlackTree.check_correct_properties = check_correct_properties
```

در عملیات حذف هم ابتدا مانند یک د.د.ج عادی عمل می‌کنیم، اگر راسی که حذف کردیم قرمز باشد که هیچ کدام از خواص درخت خراب نخواهد شد. ولی اگر سیاه باشد خاصیت پنجم خراب خواهد شد.
