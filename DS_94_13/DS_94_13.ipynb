<h4>
جستجوی دودویی
</h4>
<p></p>

<li>
مسأله: دنباله‌ای مرتب به ما داده شده است، الگوریتمی طراحی کنید که بگوید عدد دلخواه $x$ در میان این اعداد وجود دارد یا خیر.
</li>
یک راه پیشنهادی می‌تواند به این صورت باشد:
<br>
اگر آرایه را به $a$ دسته‌ی $n / a$تایی تقسیم کنیم. در $O(a)$ و تنها مقایسه با ابتدای دسته‌ها می‌توان دریافت که $x$ در کدام دسته می‌باشد (آخرین دسته‌ای که $x$ از عضو ابتدای آن بزرگتر باشد). سپس در $O(n / a)$ با مقایسه با تمامی اعداد داخل آن بازه، متوجه می‌شویم که $x$ در بین آنها موجود است یا نه.
<br>
منطقی است که مقدار $a$ را برابر با $sqrt(n)$ قرار دهیم و در $O(sqrt(n))$ می‌توانیم بگوییم $x$ در این دنباله موجود هست یا نه.

<br>
<font size=4>
(راه حل بالا ربطی به جست‌و‌جوی دودویی ندارد و دارد!)
</font>
<br>
<br>
می‌توانستیم الگوریتم بالا را با عدد $a$ ثابتی، دوباره به صورت بازگشتی بر روی دسته‌ی پیدا شده انجام دهیم. در این صورت تحلیل زمانی آن به صورت رابطه‌ی بازگشتی زیر می‌شود:
<br>
$T(n) = a + T(n / a)$
<br>
در این حالت عدد مناسب می‌تواند $a = 2$ باشد.
<br>
$T(n) = 2 + T(n / 2) = \theta(log n)$
<br>
<font size=4>
(راه حل بالا ربطی به جست‌و‌جوی دودویی دارد!)
</font>
<br>
<br>

<b>
جست‌و‌جوی دودویی:
</b>
<br>
این الگوریتم برای یافتن عنصر خاصی به کار می‌آید.
<br>
در هر مرحله فرض می‌کنیم جواب در بازه‌ی $[l, r)$ باشد (ابتدا کل شامل آرایه است).
<br>
در هر مرحله $mid$ را برابر $(l+r)/2$ قرار می دهیم.
<br>
اگر جواب مورد نظر قبل از $mid$ بود جواب را در بازه $[l, mid)$ جست و جو می کنیم.
<br>
در غیر اینصورت جواب را در بازه ی $[mid, r)$ جست و جو می کنیم.
<br>
با توجه به اینکه طول بازه در ابتدا برابر $n$ است و  در هر مرحله طول آن نصف می شود مرتبه زمانی الگوریتم از $theta(logn)$ است.
<br>
$T(n)=T(n/2)+1 =\theta(logn)$
<br>
مسأله‌ی وجود عدد $x$ در آرایه‌ی $a$ را پیاده‌سازی می‌کنیم.
</font></div>

```python
def binary_search(a, x):
    begin, end = 0, len(a)
    
    # current interval = [begin, end) -> len = end - begin
    while end - begin > 1: # it must be splitable
        mid = (begin + end) / 2
        
        if a[mid] > x:
            end = mid
        else: # x >= a[mid]
            begin = mid
    
    return a[begin] == x
    

a = [1, 3, 3, 5, 7, 9, 9, 11, 13]

for i in range(5):
    print "Is %d in array? %s" % (i, binary_search(a, i))
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<p></p>
در انتهای جست‌و‌جوی دودویی یکی از اشاره‌گرهای begin و end جواب مورد نظر ما را در خود دارند.
اما کدام؟
<br>
پیاده‌سازی <b> اشتباه </b> پایین برای یافتن اولین عدد بزرگتر مساوی $x$ در آرایه‌ی $a$ می‌باشد.
</font></div>

```python
def upper_bound_wrong(a, x):
    begin, end = 0, len(a)
    
    while end - begin > 1:
        mid = (begin + end) / 2
        if a[mid] >= x:
            end = mid
        else:
            begin = mid
            
    if end == len(a):
        print "%d was greater than all" % x
    
    return end

a = [1, 3, 3, 5, 7, 9, 9, 11, 13]
print upper_bound_wrong(a, 1)
print upper_bound_wrong(a, 2)
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<p></p>
در مثال بالا به ازای عدد ۱، اولین عدد بزرگتر مساوی آن را ۳ (در اندیس شماره ۱) می‌دهد اما جواب اصلی عدد ۱ در اندیس ۰ است.
<br>
برای رفع این ابهامات روشی ارائه می‌دهیم:
اگر شرط مقایسه‌ی جست‌و‌جوی دودویی را تابعی مانند $f(i)$ در نظر بگیریم که ۰ یا ۱ بر می‌گرداند، دنباله‌ی $f(0), f(1), ... , f(i), ... , f(n - 1)$ تعدادی ۰ در ابتدا دارد و از جایی به بعد ۱ می‌شود (به صورت $0^k1^m$ می‌باشد به شرطی که $k + m = n$. با این توصیف جست‌و‌جوی دودویی کاری می‌کند تا مرز ۰ و ۱ را بگونه‌ای سریع پیدا کنیم.
<br>

روش ما برای رفع این ابهامات به صورت زیر می‌باشد:
همیشه اشاره‌گر end به ۱ در دنباله‌ی $f$ (که بالا تعریف شد) و begin به ۰ اشاره کند.
<br>
از آنجا که ما بازه‌هایمان را به صورت بسته باز $[begin, end)$، end در ابتدا جایی خارج از خانه‌های مجاز است. پس می‌توان فرض کرد end در ابتدا به ۱ در دنباله‌ی $f$ اشاره می‌کند و اگر درست عمل کنیم end همیشه در ۱‌ها باقی می‌ماند.
<br>
اما اگر ممکن باشد $f$ تماما ۱ باشد، پس begin نباید به خانه‌ی اول اشاره کند در ابتدای کار. یک راه حل می‌تواند این باشد که در همان ابتدا چک کنیم خانه‌ی begin آیا ۱ هست یا نه و خروجی مد نظر را بدهیم. راه دیگر آن است که begin را یک خانه عقب‌تر (مثلا خانه‌ای غیرواقعی فرض کنیم) و $f$ آن حالت را ۰ در نظر بگیریم.
<br>
برای مثال کد قبلی ما تابع f را برابر با $a[i] >= x$ گذاشتیم و به دنبال اولین ۱ بودیم و چون end قرار است به ۱ اشاره کند، خروجی end می‌شود، اما end نمی‌تواند از جای اولیه‌ی begin برسد، برای همین به ازای بعضی داده‌ها اشتباه عمل می‌کند.
<br>
در مثال بالا که ممکن است تابع تماما ۱ باشد، begin را به خانه‌ی -۱ در ابتدا اشاره می‌دهیم و $f$ خانه‌ی فرضی را ۰ فرض می‌کنیم.
</font></div>

```python
def upper_bound(a, x):
    begin, end = -1, len(a)
    
    while end - begin > 1:
        mid = (begin + end) / 2
        if a[mid] >= x:
            end = mid
        else:
            begin = mid
            
    if end == len(a):
        return None 
    return end

for i in range(5):
    print "Index of the first item >= %d is %s" % (i, upper_bound(a, i))
```

<h4>
یافتن k-امین عنصر در زمان خطی
</h4>
<p></p>

مسأله‌ای که به آن می‌پردازیم، یافتن k-امین عنصر در آرایه‌ای دلخواه است (عدد k-ام در دنباله‌ی مرتب شده آن). یک راه ساده‌ی $O(n log n)$ می‌تواند مرتب کردن آن و پیدا کردن k-امین عضو آن باشد. الگوریتمی بهتر از الگوریتم‌های خطی نمی‌توان ارائه داد، چون مسأله‌ی پیدا کردن عضو کمینه (عضو ۰-ام) را نمی‌توان از $o(n)$ برای دنباله‌ی دلخواه حل کرد! در ادامه در تلاش برای یافتن الگوریتم سریعتر برای مسأله هستیم. ۲ روش را توضیح می‌دهیم:

<ul>
<li>
جست‌و‌جوی تصادفی
</li>
<li>
الگوریتمی معروف به میانه‌ی میانه‌ها
</li>
</ul>
<br>

<h4>
جست‌و‌جوی تصادفی
</h4>
<br>

در این روش همانند مرتب‌سازی سریع، در هر مرحله، عضوی 
<b>
تصادفی
</b>
مانند x را انتخاب کرده و آرایه را بر اساس آن به ۳ دسته تقسیم می‌کنیم :
<ol>
<li>
اعداد کوچکتر از x
</li>
<li>
اعداد برابر با x
</li>
<li>
اعداد بزرگتر از x
</li>
</ol>

با توجه به اندازه‌ی دسته‌ها و عدد k، این الگوریتم را به صورت بازگشتی در دسته‌ی مورد نظر انجام می‌دهیم (البته در صورتی که در دسته‌ی مساوی با x باشد ادامه نمی‌دهیم).
<br>
واضح است در بدترین حالت ممکن است برنامه $O(n ^ 2)$ به طول بینجامد اما می‌توان اثبات کرد در حالت میانگین پیچیدگی زمانی از مرتبه‌ی $O(n)$ دارد.
<br>
۲ مسأله:
<ol>
<li>

اگر دسته‌ی برابر با x را جدا نمی‌کردیم و تنها عملیات pivot همانند مرتب سازی سریع، انجام می‌دادیم. سپس الگوریتم را به طور بازگشتی بر روی $[0, i)$ یا $[i + 1, n)$ اجرا می‌کردیم، ممکن بود مشکلی پیش بیاید؟ (دقت کنید در هر مرحله دست کم عضوی که pivot شده است، در هیچ کدام نمی‌آید در نتیجه پایان پذیر است. به اعداد مساوی دقت کنید!)

</li> 
<li>
ثابت کنید در حالت میانگین به چه دلیل $O(n)$ می‌شود. (یا تحقیق کنید!)
</li>
</ol>
<br>
در ادامه پیاده‌سازی از الگوریتم بالا ارائه می‌دهیم.
خوب هست که جست‌و‌جو کنید و شبه کد‌های دیگر از جمله کتاب CLRS را ببینید. پیاده سازی ارائه شده کمی متفاوت هست.
</font></div>

```python
import random

def random_select(a, k):
    assert len(a) > k

    v = a[random.randint(0, len(a) - 1)]

    # less than, equal, greater than x
    lt = [x for x in a if x < v]
    eq = [x for x in a if x == v]
    gt = [x for x in a if x > v]

    if len(lt) > k:
        return random_select(lt, k)
    elif len(eq) + len(lt) > k:
        return v
    else:
        return random_select(gt, k - len(eq) - len(lt))

a = [5, 6, 0, 1, 10, 12, -1, 8, 100]

print random_select(a, 0)
print random_select(a, 2)
print random_select(a, 5)
print random_select(a, 8)
```

<h4>
الگوریتم میانه‌ی میانه‌ها
</h4>
<p></p>
الگوریتم به صورت زیر می‌باشد:

<ul>
<li>
آرایه را به $n/5$ دسته به اندازه‌ی ۵ تقسیم می‌کنیم.
</li>
<li>
هر دسته را مرتب کرده و میانه‌ی آن را پیدا می‌کنیم.
</li>
<li>
به صورت بازگشتی میانه‌ی میانه‌های $n / 5$ دسته‌ی یاد شده را انتخاب می‌کنیم.
</li>
<li>
فرض کنید $x$ میانه ای باشد که در مرحله ۴ بدست آورده ایم. حال تمام دسته هایی که میانه آن ها از $x$ کمتر است را در نظر بگیرید تعداد آنها $n / 10$ است.

در هر یک از این دسته ها دو عضو کوچکتر از میانه خود دسته به همراه میانه انتخاب کرده از آن دسته مقداری کمتر از $x$ دارند و تعداد آن ها $3 n / 10$ است.
به طور مشابه می توان ثابت کرد حداقل $3n / 10$ عنصر بزرگتر از $x$ در آرایه وجود دارد.
</li>
<li>

حال می توانیم آرایه را براساس $x$ به دو دسته اعداد کوچکتر از $x$ و اعداد بزرگتر از $x$ تقسیم کنیم و عدد مورد نظر را در دسته مورد نظر به صورت بازگشتی به کمک الگوریتم گفته شده 
پیدا کنیم.
</li>
</ul>

<img src="images/ithselect.png" width="60%">

با توجه به اینکه دسته اعداد کوچکتر از $x$ و دسته اعداد بزرگتر از $x$ حداکثر $7n/10$ عضو دارند مرتبه زمانی الگوریتم فوق در بدترین حالت از فرمول زیر بدست می آید:

$T(n)<=T(n/5)+T(7n/10)+c.n$

که $T(n/5)$ زمان لازم برای پیدا کردن میانه میانه هاست
و $T(7n/10)$ زمان لازم برای حل مساله در مرحله ی بعد بازگشتی است.

از فرمول بالا می توان با استقرا دریافت که 
$T(n)$ از مرتبه زمانی $\theta(n)$ است.
<br>
پرسش:
<br>
فرض کنید در الگوریتم میانه ی میانه ها به جای دسته های ۵ تایی دسته ها را ۳تایی می کردیم. آیا بازهم الگوریتم از $O(n)$ باقی می ماند؟ ۷تایی چطور؟
<br>
راه حل:
در حالت ۳تایی تابع بازگشتی بصورت زیر می شود:
<br>
$T(n)=T(n/3)+T(4n/6)+c.n = \theta(nlogn)$
<br>
و در حالت 7تایی تابع بازگشتی بصورت زیر می شود:
<br>
$T(n)=T(n/7)+T(10n/14)+c.n =\theta(n)$
<br>
