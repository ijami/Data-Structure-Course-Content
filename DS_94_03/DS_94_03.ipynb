</font></div>

مثال
</font>
تابع های زیر را با هم مقایسه می کنیم
<br>
$$1/6N^3 + 20N + 16$$

$$1/6N^3 + 100N^{4/3} + 56$$

$$1/6N^3 - 1/2N^{2} + 1/3N$$

$$1/6N^3 $$

برای این کار ابتدا یک تابع $plot1(max_n)$ تعریف می‌کنیم که نمودار سه تابع فوق را در بازه‌ی ۱ تا $max_n$ رسم می‌کند. 
</font></div>

```python
def plot1(max_n):
    figure(figsize=(20, 10))
    xlabel("N")
    x = range(1, max_n+1)

    def f(n): return 1.0/6*n**3 + 20*n + 16
    y = [f(n) for n in x]
    plot(x, y, 'r', label='1/6n^3 + 20n + 16')

    def f(n): return 1.0/6*n**3 + 100*n**(4.0/3) + 56
    y = [f(n) for n in x]
    plot(x, y, 'g', label='1/6n^3 + 100n^(4/3) + 56')

    def f(n): return 1.0/6*n**3 - 1.0/2*n**2 + 1.0/3*n
    y = [f(n) for n in x]
    plot(x, y, 'b', label='1/6n^3 - 1/2n^2 + 1/3n')

    y = [1.0/6*n**3 for n in x]
    plot(x, y, 'k', label='1/6n^3')
    
    legend(loc=2)
```

مثال
</font>
<p></p>
می توان نتیجه گرفت
<br>
$$1/6N^3 + 20N + 16 \quad \; \; \sim { 1/6N^3}$$
$$1/6N^3 + 100N^{4/3} + 56 \>   \sim 1/6N^3$$
$$1/6N^3 - 1/2N^{2} + 1/3N   \sim 1/6N^3$$
<br>
<b>
می توان از جمله ها با درجه کوچکتر از بزرگترین درجه چشم پوشی کرد.
</b>
<ul>
    <li> در N
    های کوچک معمولا مقدار توابع کم است و برای ما مهم نیست.
    </li>
    <li> در N
    های بزرگ تاثیر خیلی کمی روی تابع می گذارند.
    </li>
</ul>
به عبارت دیگر در این مثال‌ها هر دو تابع $f$ و $g$
روند رشد یکسانی دارند چرا که:
$$ \lim_{n\to\infty}\frac{f(n)}{g(n)} = 1 \to f(n) \sim g(n) $$
</font></div>

<font color=#FF7500>
مثال
</font><p></p>
می‌خواهیم تعداد عملیات‌های کد ساده‌ی زیر را محاسبه کنیم.
<br />

<p></p>

</font></div>

```python
def algorithm1(n):
    i = 0
    while i < n:
        print(i)
        i += 1
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
$c_i$: تعداد عملیات‌هایی که CPU برای انجام خط $i$ ام نیاز دارد
<br />
با توجه به تعریف بالا، تعداد عملیات‌های تابع algorithm1 برحسب n چقدر است؟
<br />
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
$m(n)$: تعداد دقیق عملیات‌ها بر حسب $n$ $\Leftarrow$ $m(n) = c_1 + n \times (c_2 + c_3 + c_4)$
<br />
$f(n)$: تعداد عملیات‌ها اگر به‌جای تمام $c_i$ ها کمترین $c_i$ را در نظر بگیریم.
<br />
$g(n)$: اگر به جای تمام $c_i$‌ها بیشترین $c_i$ را در نظر بگیریم.
<br />
$h(n)$: اگر به جای تمام $c_i$‌ برابر با یک باشد.
<br />
عبارات زیر را محاسبه کنید:
$$\lim_{n\to\infty}{\frac{m(n)}{f(n)}}$$
<br />
$$\lim_{n\to\infty}{\frac{f(n)}{g(n)}}$$
<br />
$$\lim_{n\to\infty}{\frac{g(n)}{h(n)}}$$
<br />
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<font color=#FF7500>
حالت‌های میانگین، بهترین و بدترین
</font><p></p>
تعداد دفعات اجرای یک خط می‌تواند به بسیاری از موارد و به طور کلی به تمامی عناصر ورودی مرتبط باشد. اما برای مقایسه‌ی زمان اجرای الگوریتم‌ها تعداد دقیق دفعات اجرا مورد نظر نیست و به همین علت در بررسی زمان اجرا معمولا یک یا چند معیار مهم را مانند تعداد عناصر ورودی به عنوان متغیر در نظر گرفته و سپس تعداد عملیات الگوریتم را نسبت به این متغیرها در یکی از سه حالت زیر مورد بررسی قرار می‌دهند:

</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<ol>
<li>
بهترین حالت(Best Case):
اگر اندازه‌ی ورودی الگوریتم را $n$ در نظر بگیریم، بهترین حالت وقتی است که یک ورودی به اندازه‌ی $n$ از همه‌ی ورودی‌های هم‌اندازه‌ی خود زمان اجرای کمتری داشته باشد.
</li>
<li>
بدترین حالت(Worst Case):
برعکس حالت قبل، نوعی از ورودی به اندازه‌ی $n$ که از همه‌ی ورودی‌های هم‌اندازه‌ی خود زمان اجرای بیشتری نیاز داشته‌باشد.
</li>
<li>
حالت میانگین(Average Case):
زمان اجرای الگوریتم وقتی که یکی از همه‌ی ورودی‌های ممکن با اندازه‌ی $n$ به طور شانسی (با احتمال یکسان) به برنامه داده‌شود. یا به عبارت دیگر، متوسط زمان اجرای برنامه برای همه‌ی ورودی‌های به اندازه‌ی $n$.
</li>
</ol>

<ol>
<li>
بهترین حالت(Best Case):
<b>حد پایینی از زمان اجرا</b>
</li>
<li>
بدترین حالت(Worst Case):
<b>حد بالایی از زمان اجرا</b>
</li>
<li>
حالت میانگین(Average Case):
<b>زمان اجرا به ازای ورودی تصادفی(Random)</b>
</li>
</ol>

در این جلسه تنها دو روش بهترین حالت و بدترین حالت را بررسی می‌کنیم. 


</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<font color=#FF7500>
مثال
</font><p></p>
کد زیر یک پیاده‌سازی از الگوریتم مرتب‌سازی حبابی است که در جلسه‌ی قبل مطرح شد. می‌خواهیم تعداد عملیات‌های این الگوریتم در بهترین و بدترین حالت را محاسبه کنیم.
<p></p>


</font></div>

```python
def bubble_sort(A):
    n = len(A)
    for i in range(n - 1):
        for j in range(n - 1):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]
    return A
```

<font size=3> * توجه کنید که شرط حلقه یک بار بیشتر از دفعاتی که وارد حلقه می‌شویم بررسی می‌شود. </font>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<p></p>
فرض کنید $a_i$ مجموع تعداد عملیات CPU باشد که در طول اجرای الگوریتم بر روی سطر $i$ ام از کد بالا اجرا می‌شود. 
می‌خواهیم ببینیم
تعداد دقیق عملیات‌ها بر حسب $n$ در بدترین و بهترین حالت چقدر است؟
<br />
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
در بهترین حالت داریم:
$$a_1 + n * a_2 + (n - 1) * n * a_3 + (n - 1) * (n - 1) * a_4 + 0 * a_5 + a_6$$
در بدترین حالت داریم:
$$a_1 + n * a_2 + (n - 1) * n * a_3 + (n - 1) * (n - 1) * a_4 + (n - 1) * (n - 1) * a_5 + a_6$$
توابعی شبیه توابعی که برای الگوریتم قبل در نظر گرفته‌ایم تعریف کنید. حد نسبت این توابع به یک‌دیگر نیز مانند توابع بالا یک عدد ثابت خواهد بود.
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
می‌خواهیم تعداد عملیات‌های الگوریتم مرتب‌سازی حبابی را با الگوریتم ابتدایی که اعداد $0$ تا $n - 1$ را چاپ می‌کرد مقایسه کنیم. برای سادگی فرض کنید که 
$C = max(a_i, b_i, c_i)$
. در این‌صورت می‌توان کران بالای زیر را در بدترین حالت برای الگوریتم مرتب سازی حبابی در نظر گرفت:
$$T1(n) = C * (3n^2 - 5n + 4)$$
و همچنین کران بالای زیر برای الگوریتم ابتدایی به دست می‌آید:
$$T2(n) = C * (3n + 1)$$
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
حالا می‌خواهیم زمان اجرای مرتب‌سازی حبابی را محاسبه کنیم.
</font></div>

```python
from random import randrange
```

```python
def get_time(n):
    global nums # should be global to access inside timeit
    nums = [randrange(1000 * 1000) for i in range(n)]
    res = %timeit -n1 -r1 -o bubble_sort(nums)
    return res.best
```

```python
N = [1, 10, 100, 200, 500, 1000, 2000, 4000, 8000, 16000]
time = [0] * len(N)
for i in xrange(len(N)):
    print "Estimating running time for N =",N[i]
    time[i] = get_time(N[i])
```

```python
figure(figsize=(20, 20))
plot(N, time)
```

```python
def plot2(max_n):
    figure(figsize=(20, 20))
    xlabel("n")
    ylabel("Time")
    C = 1.5
    max_n = 100 
    x = range(1, max_n)
    y = [C * (3 * n + 1) for n in x]
    plot(x, y, 'r', label='T1(n)=C * (3n + 1)')
    y = [C * (3 * (n * n) - 5 * n + 4) for n in x]
    plot(x, y, 'b', label='T2(n)=C * (3n^2 - 5n + 4)')
    y = [n for n in x]
    plot(x, y, 'g', label='n')
    y = [3*(n * n) for n in x]
    plot(x, y, 'y', label='3*n^2')
    legend(loc=2)
    
plot2(100)
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
همان‌طور که در نمودارهای بالا مشاهده می‌شود با افزایش $n$، کران بالای مرتب‌سازی حبابی نردیک تابع $n^2$ می‌ماند و کران بالای الگوریتم ابتدایی نزدیک تابع $n$ می‌ماند اما این دو دسته به وضوح از یکدیگر دور می‌شوند. به شکل دقیق‌تر می‌توان گفت $\lim_{n\to\infty}\frac{T1(n)}{n^2}$ و $\lim_{n\to\infty}\frac{T2(n)}{n}$ هر یک برابر با عددی ثابت می‌شوند درحالی‌که $\lim_{n\to\infty}\frac{T1(n)}{T2(n)}$ یک چندجمله‌ای خطی است که با افزایش $n$ بی‌کران بزرگ می‌شود. 
<br />
بنابراین با وجود آن‌که تابع کران بالای مرتب‌سازی حبابی و تابع $n^2$ متفاوت هستند، به نظر می‌آید در $n$های به اندازه‌ی کافی بزرگ، تفاوت اصلی این توابع بین دو دسته از آن‌هاست که یک دسته نزدیک تابع $n$ باقی می‌ماند و دسته‌ی دیگر نزدیک تابع $n^2$ باقی می‌ماند. حال می‌توان گفت که بنابراین در مقایسه‌ی زمان الگوریتم‌ها بین دو الگوریتم از یک دسته تفاوت چندانی وجود ندارد و دو الگوریتم زمانی به طور قابل توجه از لحاظ زمانی متفاوت هستند که از دو دسته‌ی متفاوت باشند.
<br />
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
حال لازم است که تعریفی دقیق‌تر برای مفهوم دسته ارائه کنیم. برای این‌کار از مجموعه‌های زیر استفاده می‌کنیم:
$$\mathcal{O}(g(n)) = \{f(n) ~ | ~ \exists c, n_0: 0 \le f_n \le cg(n) ~ \forall n > n_0\}$$
$$\Omega(g(n)) = \{f(n) ~ | ~ \exists c, n_0: 0 \le cg(n) \le f(n) ~ \forall n > n_0\}$$
$$\Theta(g(n)) = \{f(n) ~ | ~ f(n) \in O(g(n)) ‍~ and ~ f(n) \in \Omega(g(n))\}$$

در واقع اگر یک تابع $f(n)$ در $\mathcal{O}(g(n))$ باشد به عبارت نادقیق می‌توان گفت که این تابع برای $n$های به اندازه‌ی کافی بزرگ، $f(n) \le g(n)$ و هم‌چنین درمورد $f(n) \in \Omega(g(n))$ می‌توان به صورت شهودی عبارت $f(n) \ge g(n)$ را به کار برد. با توجه به این دو عبارت اگر $f(n) \in \Theta(g(n))$ درواقع $f(n)$ و $g(n)$ در یک دسته قرار دارند. 
<br />
برای بررسی توابع، سعی می‌کنیم تابع ساده‌تر و هم‌دسته‌ی آن را پیدا کنیم. برای الگوریتم ابتدایی این تابع $f(n) = n$ و برای الگوریتم مرتب سازی حبابی این تابع همان $n^2$ است. در هنگام یافتن تابع ساده‌تر از ضرایب صرف نظر می‌کنیم و بزرگ‌ترین عنصر در تابع اولیه را در نظر می‌گیریم.

تابع f در کد زیر در بدترین حالت در چه زمانی اجرا می شود؟
<br>
تابعی هم‌دسته با تابع زمان اجرای $f$ پیدا کنید. ساده‌ترین این توابع کدام هستند؟ 
</font></div>

```python
def f(N):
    sum = 0;
    for i in range(N):
        for j in range(i+1, N):
            for k in range(j+1, N):
                for h in range(k+1, N):
                    sum += 1
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
زمان اجرای تابع g
در کد زیر را نیز بررسی کنید.
</font></div>

```python
def G(N, x, A): #A: array of size N
    s, e = 0, N
    while(e - s > 1):
        mid = (s+e)//2
        if (A[mid] <= x):
            s = mid
        else:
            e = mid
    return A[s] == x
```

```python
A = [1, 5, 10, 22, 31, 44, 55, 69, 75, 85, 100]
N = len(A)
G(N, 78, A)
G(N, 85, A)
```

```python
figure(figsize=(20, 20))
xlabel("n")
ylabel("val")
ylim(0, 800)
max_n = 50
x = range(1, max_n + 1)
y = [1 for n in x]
plot(x, y, 'r', label='1',linewidth=5)
y = [math.log(n) for n in x]
plot(x, y, 'g', label='log(n)')
y = [n for n in x]
plot(x, y, 'b', label='n')
y = [n * math.log(n) for n in x]
plot(x, y, 'y', label='n * log(n)')
y = [n ** 2 for n in x]
plot(x, y, 'c', label='n^2')
y = [2 ** n for n in x]
plot(x, y, 'k', label='2^n')
y = [math.factorial(n) for n in x]
plot(x, y, 'm', label='n!')
legend(loc=1)
```
