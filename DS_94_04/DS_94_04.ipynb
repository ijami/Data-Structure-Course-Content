قرارداد: با توجه به این که واژه‌ی تابع در برنامه‌نویسی و ریاضی دو مفهوم نزدیک به هم اما متفاوت دارد، در سرتاسر این فصل از تابع به معنای ریاضی آن استفاده می‌شود و هرگاه بخواهیم از یک تابع برنامه صحبت کنیم، از واژه‌های روند یا رویه استفاده می‌کنیم. هم‌چنین در این متن هر جا دامنه‌ی یک تابع مشخص نشده است، اعداد طبیعی مد نظر هستند و همه‌ی ثابت‌ها اعداد حقیقی مثبتند.

<p></p>
بسیاری از الگوریتم‌هایی که ما در این درس با آن‌ها سر و کار داریم بر اساس تکرار بازگشتی یک یا چند رویه کار می‌کنند. برای مثال مرتب‌سازی ادغامی در زمان اجرا بر روی یک آرایه‌، به صورت بازگشتی خودش را روی دو نیمه‌ی ابتدا و انتهای آرایه فراخوانی کرده و سپس دو نیم‌آرایه‌ی مرتب حاصل را ادغام می‌کند. تحلیل زمان اجرای این نوع الگوریتم‌ها کمی پیچیده‌تر از الگوریتم‌های سرراستی مثل مرتب‌سازی شمارشی یا حبابی است چرا که در زمان تحلیل آن‌ها با توابع بازگشتی سر و کار پیدا می‌کنیم.
<p></p>
به عنوان اولین و ساده‌ترین نمونه بیایید به جست و جوی دودویی دقت کنیم. فرض کنید یک آرایه‌ی مرتب با 
$n$
عنصر به شما داده شده است و سپس یک نفر مرتبا از شما سؤال‌هایی به این فرم می‌پرسد:
آیا عدد 
$x$
در این آرایه ظاهر شده است؟

<p></p>
هدف شما این است که به این سؤال‌ها در کوتاه‌ترین زمان ممکن پاسخ دهید. کمی بعدتر در همین درس با تکنیک درهم‌سازی آشنا خواهید شد که می‌تواند الگوریتم‌های جالب‌تری برای این کار به شما بدهد اما در حال حاضر از این الگوریتم که نام آن جست و جوی دودویی است استفاده می‌کنیم:

<p></p>
هر بار عنصر وسط آرایه را انتخاب می‌کنیم. اگر این عنصر برابر با همان $x$ 
بود که به دنبال آن می‌گردیم، جواب سؤال پیدا شده است، در غیر این صورت اگر این عنصر از هدف ما کوچک‌تر بود فقط در نیمه‌ی بعد از عنصر وسط به جست و جو ادامه می‌دهیم و در غیر این صورت فقط در نیمه‌ی قبل از عنصر وسط. این روند را می‌توان به این شکل پیاده‌سازی کرد:

<p></p>
(در این‌جا $up$ و $down$ انتها و ابتدای بازه‌ای هستند که در حال جست و جو در آنیم.)

</font></div>

```python
def binary_search(array, x):
    up = len(array)-1
    down = 0
    while up>=down:
        mid = (up+down)//2
        if array[mid]==x:
            return "FOUND"
        elif array[mid]<x:
            down=mid+1
        else:
            up=mid-1
    return "NOT FOUND"
```

```python
print (binary_search([1, 2, 3, 4, 5, 6, 100, 110], 3))
print (binary_search([1, 2, 3, 4, 5, 6, 100, 110], 7))
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>

<hr>

برای این که ماهیت بازگشتی این روند بهتر مشخص شود، آن را به این صورت به شکل بازگشتی پیاده‌سازی می‌کنیم:
</font></div>

```python
def recursive_binary_search(array,x,up,down):
    if down>up:
        return "NOT FOUND"
    mid = (up+down)//2
    if array[mid]==x:
        return "FOUND"
    elif array[mid]<x:
        return recursive_binary_search(array,x,up,mid+1)
    else:
        return recursive_binary_search(array,x,mid-1,down)
```

```python
array = [1, 2, 3, 4, 5, 6, 100, 110]
print(recursive_binary_search(array,3, len(array)-1, 0))
print(recursive_binary_search(array,7, len(array)-1, 0))
```

حال بیایید زمان اجرای این رویه‌ی بازگشتی را به دست آوریم. فرض کنید زمان اجرای این رویه در بدترین حالت (حالتی که بیش‌ترین زمان مصرف می‌شود) روی یک آرایه با اندازه‌ی 
$n$
را با 
$T(n)$
نمایش دهیم. 

می‌خواهیم کلاس پیچیدگی تابع $T(n)$ را به دست آوریم. 

<p></p>
اولین روشی که استفاده می‌کنیم یکی از روش‌های کلاسیک در علوم کامپیوتر به حساب می‌آید و حتی در کارهای قدیمی آلن تورینگ هم به چشم می‌خورد. این روش بیش‌تر برای نشان‌دادن پایان‌پذیربودن یک رویه کاربرد دارد اما گاهی، مثل حالا، می‌تواند در به دست آوردن زمان اجرای آن نیز کارآمد باشد. 

<p></p>
 یکی از خواص به نظر بدیهی اعداد صحیح، اصل خوشترتیبی است که بیان می‌کند که هر زیرمجموعه‌ی ناتهی از اعداد صحیح که از پایین کراندار باشد
 یک کوچک‌ترین عنصر دارد.
 ما از یک بیان دیگر این اصل، که به اصل نفی نزول نامتناهی معروف است، و بیش‌تر به کارمان می‌آید استفاده می‌کنیم. (هر دوی این اصل‌ها با اصل استقرا هم‌ارزند.)
 <p></p>
 اصل نفی نزول نامتناهی (نسخه‌ی کامپیوتری!) : 
 اگر یک متغیر صحیح نامنفی به نام $i$ داشته باشیم که در هر بار فراخوانی یک رویه‌ی بازگشتی مقدار $i$ کاهش یابد، آن‌گاه آن رویه همواره پایان می‌پذیرد و اگر امکان این که $i$  در دو فراخوانی بازگشتی مختلف این رویه مقدار یکسانی بگیرد وجود نداشته باشد، آن‌گاه تعداد دفعات اجرای این رویه حداکثر $i$ بار است.
 
 <p></p>
 چیزی که در بالا آمده حالت کلی این استدلال است: فرض کنید در جست و جوی دودویی متغیر طول آرایه را به عنوان 
 $i$ در تعریف بالا در نظر بگیریم.
 در هر بار فراخوانی بازگشتی این رویه طول بازه‌ای که در آن به جست و جو می‌پردازیم کاهش می‌یابد، پس امکان ندارد جست و جو تا ابد ادامه پیدا کند. (جست و جو قطعا پایان می‌یابد) علاوه بر این واضح است که در یک فرآیند جست و جو هیچ‌وقت اتفاق نمی‌افتد که به یک مقدار تکراری برای طول آرایه برسیم، در نتیجه رویه‌ی 
 $recursive\_binary\_search$
 حداکثر $i$ بار فراخوانی می‌شود. که چون در ابتدای کار $i$ برابر با طول آرایه یا همان $n$ بود و هر بار فراخوانی این رویه یک زمان ثابت مصرف می‌کند می‌توانیم نتیجه بگیریم که فرآیند جست و جو در مجموع از
 $O(n)$ 
 است.
 
 <p></p>
 اما انتخاب بهتری برای $i$ هم وجود داشت. به راحتی می‌توانیم $i$ را برابر با لگاریتم طول بازه‌ای که در آن جست و جو می‌کنیم تعریف کنیم. این بار مشاهده می‌کنیم که باز هم $i$ با هر فراخوانی بازگشتی رویه کاهش می‌یابد و امکان گرفتن مقادیر تکراری را هم ندارد. در نتیجه زمان اجرای این الگوریتم از 
 $O(\log n)$ 
 است.
 (در این‌جا دیگر $i$ یک عدد صحیح نبود ولی اگر جزء صحیح آن را در نظر بگیریم آن‌گاه مشکل حل است.)
 
 <p></p>
 آیا می‌توانید نشان دهید که این الگوریتم در بدترین حالت از 
 $\Omega(\log n)$
 است و در نتیجه از 
 $\Theta(\log n)$
 می‌باشد؟ این کار را می‌توانید با ساختن بی‌نهایت مثال انجام دهید! 
 
 <p></p>
 تذکر: ما همیشه در مبنای ۲ لگاریتم می‌گیریم.

<font color=#FF7500>
روش حدس و استقرا
</font><p></p>
کاری که در این‌جا برای تحلیل زمان اجرای جست و جوی دودویی انجام دادیم بسیار سخت بود. اگر بخواهیم الگوریتم‌های بازگشتی را این‌طور تحلیل کنیم باید همیشه به دنبال متغیر مناسب بگردیم و همان‌طور که در مثال بالا مشخص است، پیدا کردن متغیرهای دیگر می‌تواند جواب‌های درست ولی نه چندان بهینه به ما بدهد. راستش را بخواهید تا پایان این قسمت با هیچ روش غیرحدسی برخورد نخواهید کرد، اما تکنیک‌های متفاوتی را خواهید دید که به شما کمک می‌کند بهتر حدس بزنید یا برای اثبات حدستان کم‌تر مشقت بکشید. البته این تکنیک‌ها به ترتیب از ساده به سخت (از نظر من) ارائه می‌شوند و در تحلیل یک الگوریتم در دنیای واقعی (دنیای خارج از دانشگاه!) بهترین راهکار این است که از ساده‌ترین روش‌ها استفاده کنید و هرگاه آن‌ها جواب ندادند به روش‌های قوی‌تر متوسل شوید.

<p></p>
اولین قدم ما آن است که از کد فاصله گرفته و برای تحلیل زمان الگوریتممان این زمان را به شکل یک تابع بازگشتی ریاضی نمایش دهیم. در مورد جست و جوی دودویی اگر فرض کنیم زمان لازم برای این جست و جو بر اساس تعداد مقایسه‌هایی که بین عناصر می‌کنیم تعیین می‌شود و 
$T(n)$
را برابر با تعداد مقایسه‌ها در اجرای یک نمونه جست و جوی دودویی روی یک آرایه به طول $n$ در بدترین حالت 
(حالتی که بیش‌ترین مقایسه را لازم دارد) تعریف کنیم
،می‌بینیم که 
$T(1) = 1$
و برای هر 
$n \geq 2$
داریم:
$T(n) = T(\left \lceil \frac{n-1}{2} \right \rceil) + 1$.

<p></p>
از این پس فقط به بررسی شیوه‌های پیدا کردن پیچیدگی توابعی می‌پردازیم که به شکل بازگشتی تعریف شده‌اند و به این که این توابع زمان اجرای چه رویه‌هایی را نشان می‌دهند توجه نمی‌کنیم. در این جا می‌خواهیم کلاس پیچیدگی تابع زیر را پیدا کنیم:

$$T(n)=\left\{\begin{matrix}
1 & n=1\\ 
T(\left \lceil \frac{n-1}{2} \right \rceil) + 1 & n > 1
\end{matrix}\right.$$

</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>

<hr>

باز ساده‌ترین راهی که داریم حدس زدن است! اگر خوب حدس بزنیم، می‌توانیم مسأله را حل کنیم. ما با توجه به این که جواب مسأله را می‌دانیم از قسمت قبل تقلب کرده و حدس می‌زنیم که 
$T(n) \in \Theta(\log n)$.

<p></p>
حال باید این حدس را ثابت کنیم. یکی از روش‌های بسیار کارآمد در این مواقع استفاده از استقرای ریاضی است. می‌خواهیم با استفاده از استقرای ریاضی نشان دهیم که 
$T(n) \in O(\log n)$. این معادل آن است که نشان دهیم که یک عدد ثابت 
$c$
وجود دارد که برای هر 
$n$ طبیعی به اندازه‌ی کافی بزرگ
داریم
$T(n) \leq c \log n$. 
اگر این تعریف با تعریفی که قبلا از $O$ دیده‌اید کمی متفاوت است، اندکی درنگ کنید و به خودتان بقبولانید که هر دو تعریف یک چیز را بیان می‌کنند.
<p></p>
با این حساب حالا باید دو کار انجام دهیم. اول باید یک $c$ انتخاب کنیم
و بعد باید با استقرا نشان دهیم که برای هر $n$
داریم
$T(n) \leq c \log n$. خیلی راحت می‌توانیم بگوییم بگذار $c$ یک مقدار ثابت مثلا ۱۰۰ باشد و بعد سعی کنیم با استقرا حکم را ثابت کنیم و اگر
جواب نگرفتیم، $c$ را بزرگ‌تر انتخاب کنیم و باز همین کار را ادامه دهیم و امیدوار باشیم که بالأخره به یک $c$ برسیم که کار کند. 
راه بهتر اما آن است که $c$ را انتخاب نکنیم و استقرایمان را بزنیم و وقتی که استقرا تمام شد $c$ را طوری انتخاب کنیم که روشمان جواب بدهد.  پس بیایید استقرا بزنیم. ما در این‌جا از استقرای قوی استفاده می‌کنیم.

<p></p> 
فرض استقرا آن است که به ازای هر 
$k < n$
داریم
$T(k) \leq c \log k$
و حکم استقرا آن است که 
$T(n) \leq c \log n$. حال داریم:
$$T(n) = T\left(\left \lceil \frac{n-1}{2} \right \rceil\right) + 1 \leq c \log \left \lceil \frac{n-1}{2} \right \rceil + 1 \leq c \log \frac{n}{2} + 1 \leq c \log n$$

دقت کنید که آخرین نابرابری فقط وقتی درست است که 
$ c \geq 1$ و بنابراین باید $c$ را به اندازه‌ی کافی بزرگ انتخاب کنیم.

<p></p>
حال آیا اثبات تمام شده است؟ خیر. هر استقرایی به یک پایه نیاز دارد و در این‌جا به سادگی می‌توانید ببینید که پایه‌ی استقرا برای 
$n = 1$
برقرار نیست، چرا که 
$\log 1 = 0$ اما خوش‌بختانه ما نیازی نداریم که حکم استقرا را برای همه‌ی مقادیر $n$ ثابت کنیم و کافیست برای $n$های به اندازه‌ی کافی بزرگ درست باشد. پس می‌توانیم 
$n = 2, 3$
را به عنوان پایه‌های استقرا در نظر بگیریم و $c$ را طوری انتخاب کنیم که هم در شرط بالا صدق کند و هم آن‌قدر بزرگ باشد که حکم برای این دو پایه درست باشد. آن‌گاه با استقرا حکم ثابت شده است.
(چرا نمی‌توانیم فقط $n=2$ را به عنوان پایه‌ی استقرا در نظر بگیریم؟)

<p></p>
به عنوان یک تمرین سعی کنید با روشی مشابه ثابت کنید که 
$T(n) \in \Omega(\log n)$.

</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>

<hr>

استقرا روش بسیار قدرتمندی است اما بررسی همه‌ی جزئیات آن مانند کاری که در این‌جا انجام دادیم بسیار وقت‌گیر است و خیلی وقت‌ها لازم هم نیست. مثلا فرض کنید که می‌خواهید این تابع را بررسی کنید (این تابع از مرتب‌سازی ادغامی به دست آمده است):

$$T(n) = \left\{\begin{matrix}
0 & n = 1 \\ 
T\left(\left \lfloor \frac{n}{2} \right \rfloor\right) + T\left(\left \lceil \frac{n}{2} \right \rceil \right ) + O(n) & n \geq 2
\end{matrix}\right.$$

در این‌جا منظور از 
$O(n)$
که در تعریف تابع آمده است این است که تابعی مانند $f \in O(n)$ وجود دارد که:
$$T(n) = \left\{\begin{matrix}
0 & n = 1 \\ 
T\left(\left \lfloor \frac{n}{2} \right \rfloor\right) + T\left(\left \lceil \frac{n}{2} \right \rceil \right ) + f(n) & n \geq 2
\end{matrix}\right.$$

یا به عبارت دیگر عدد ثابت مثبت $d$ وجود دارد که:
$$T(n) \leq \left\{\begin{matrix}
0 & n = 1 \\ 
T\left(\left \lfloor \frac{n}{2} \right \rfloor\right) + T\left(\left \lceil \frac{n}{2} \right \rceil \right ) + dn & n \geq 2
\end{matrix}\right.$$

بنابراین چون در این‌جا فقط یک رابطه‌ی بازگشتی برای یک کران بالا از تابع $T$ داریم، امکان بررسی $\Omega$ وجود ندارد و فقط به دنبال این هستیم که یک بررسی $O$ انجام دهیم. 
<p></p>
اگر باز مثل قسمت قبل به حدس زدن روی بیاوریم و حدس بزنیم که 
$T(n) \in O(n^2)$
یا حتی بهتر
$T(n) \in O(n \log n)$، آن‌گاه باید این حدس را با استقرا ثابت کنیم. در این جا مقداری که برای $c$ انتخاب می‌کنیم به $d$ هم وابسته خواهد بود که با توجه به این که $d$ یک عدد ثابت است مشکلی ایجاد نمی‌کند. اما چیزی که مشکل ساز است آن کف و سقف است که استدلال‌ها را سخت‌تر می‌کنند. قطعا برای ما بسیار ساده‌تر است که با این رابطه کار کنیم:
$$
T'(n) = \left\{\begin{matrix}
0 & n = 1 \\ 
2 T'\left(\frac{n}{2}\right) + O(n)  & n>1
\end{matrix}\right.$$
در نگاه اول به نظر نمی‌رسد که جواب این رابطه با رابطه‌ی قبلی فرق قابل توجهی داشته باشد و در واقعیت هم جوابشان یکیست. اگر بتوانیم این را ثابت کنیم، بعد می‌توانیم این تابع ساده‌تر را تحلیل کنیم و از جوابش استفاده کنیم. دقت کنید که این که در این رابطه از سقف‌ها و کف‌ها صرف نظر کردیم معادل این است که فرض کنیم $n$هایمان همیشه توان‌هایی از ۲ هستند. 
این نمونه‌ای از فرض‌هاییست که ما آن‌ها را فرض‌های سازگار می‌نامیم و در موقع تحلیل پیچیدگی یک تابع از آن‌ها بسیار استفاده می‌کنیم.

<p></p>
بگذارید یک تعریف دقیق از فرض‌های سازگار بدهیم. فرض کنید که در تحلیل $O$ به جای تابع بازگشتی $T$ تصمیم می‌گیرید که چند فرض اضافه کنید و 
تابع بازگشتی 
$T'$
را تحلیل کنید، در این صورت فرض‌های شما سازگار نامیده می‌شوند هرگاه 
$T \in O(T')$. به عبارت دیگر برای هر $n$ به اندازه‌ی کافی بزرگ، فرض‌های اضافه شده یا باید مقدار تابع را افزایش دهند یا اگر کاهش می‌دهند حداکثر با یک ضریب ثابت کاهش دهند.
<p></p>
در بسیاری از توابعی که ما به آن‌ها برخورد می‌کنیم فرض‌های زیر سازگار هستند و ما آن‌ها را بدون این که هر بار ثابت کنیم سازگارند به کار می‌بریم. دقت کنید که این فرض‌ها همیشه سازگار نیستند و موقع تحلیل الگوریتم‌ها همیشه باید مواظب سازگاری فرض‌هایتان باشید. 

<ul>
  <li>این فرض که $n$ توانی از ۲ یا یک عدد ثابت دیگر است</li>
  <li>این فرض که کف و سقف‌ها را می‌توان نادیده گرفت</li>
  <li>این فرض که در استقرا نیازی به چک کردن پایه نداریم و متعاقبا این که در تعریف تابع بازگشتی نیازی به بیان حالت پایه نداریم</li>
  <li> این فرض که تابع مورد بررسی صعودی است (خیلی وقت‌ها برای اثبات سازگاری این یکی باید به مسأله‌ی اصلی رجوع کنیم)</li>
  <li> این فرض که تابع مورد بحث مثبت است (یعنی هیچ‌گاه صفر نیست) </li>
</ul>

معمولا اثبات سازگاری همه‌ی این فرض‌ها تکراری و شبیه به هم است و به مسأله بستگی چندانی ندارد. یک تمرین خوب می‌تواند این باشد که ثابت کنید این فرض‌ها در مورد تابع‌هایی که تا این‌جا بررسی کرده‌ایم سازگارند و بعد سعی کنید برای هر فرض تابعی بازگشتی پیدا کنید که با آن ناسازگار باشد. وقتی این کار را بکنید می‌بینید که تا چه میزان توابعی که این‌طور هستند شکل‌های عجیبی دارند و چه‌قدر بعید است که در تحلیل الگوریتم‌های واقعی به آن‌ها برخورد کنیم اما همیشه باید به یاد داشته باشید که این توابع هم وجود دارند و اگر یک فرض ناسازگار اضافه کنید تحلیلتان غلط است.
دقت کنید که در مورد فرض چهارم در خصوص رابطه‌ای که برای مرتب‌سازی ادغامی داریم، اگر تنها اطلاعاتی که داریم همان رابطه‌ی بازگشتی شامل $O$ باشد نمی‌توانیم ثابت کنیم که تابع باید صعودی باشد.

حال به تحلیل زمان الگوریتم مرتب‌سازی ادغامی می‌پردازیم که بعد از فرض‌های صورت گرفته به این شکل در آمده است:
$$T(n) \leq 2 T(n/2) + dn$$

<p></p>
می‌خواهیم ثابت کنیم 
$T(n) \leq c n \log n$.
تنها کاری که باید انجام دهیم بررسی گام استقرا است:
$$T(n) \leq 2 T(n/2) + dn \leq 2 \times c \times \frac{n}{2} \times (\log n - 1) + dn \leq c n \log n - cn + dn \leq c n \log n$$
در این‌جا نابرابری آخر فقط وقتی درست است که $c$ را بزرگ‌تر از $d$ انتخاب کنیم.

</font></div> 

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
حال به چند تکنیک دیگر می‌پردازیم که می‌تواند در تحلیل روابط بازگشتی کارساز باشد. تکنیک‌هایی که در این‌جا مطرح می‌کنیم کم کردن جمله‌ی کم‌ارزش، قضیه‌ی ساندویچ و تغییر متغیر هستند. یک مسأله هم با استفاده از حساب دیفرانسیل و انتگرال حل می‌کنیم. در آخر به قضیه‌ی اصلی می‌پردازیم که خلاصه‌اش این است که برای یک خانواده‌ی خیلی کاربردی از توابع بازگشتی، قبلا یک نفر دیگر استقرا را انجام داده و شما می‌توانید به راحتی از پاسخش استفاده کنید. شاخه‌ای از ریاضیات به نام ترکیبیات آنالیزی تکنیک‌های پیشرفته‌تری برای تحلیل توابع بازگشتی ارائه می‌دهد که در صورت علاقه می‌توانید آن‌ را جست و جو کنید.

<p></p>
<font color=#FF7500>
کم‌کردن جمله‌ی کم‌ارزش
</font><p></p>
به عنوان اولین مسأله فرض کنید که تابع بالا را کمی تغییر داده‌ایم و حالا به این شکل است:
$$T(n) \leq 2 T(n/2) + O(n \lg \lg n)$$
در نگاه اول مشکلی وجود ندارد. تابع را به این فرم می‌نویسیم:
$$T(n) \leq 2 T(n/2) + d n \lg \lg n$$
و سعی می‌کنیم استقرا بزنیم و ثابت کنیم که 
$T(n) \leq c n^2$. این کار ساده است:
$$T(n) \leq 2 T(n/2) + d n \log \log n \leq c n^2/2 + d n \log \log n \leq cn^2$$
انتخاب $c$ طوری که نابرابری آخر درست باشد آسان است. پس نشان دادیم که 
$T(n) \in O(n^2)$. اما آیا می‌توانیم بهتر از این کار کنیم؟ مثلا آیا می‌توانیم نشان دهیم 
$T(n) \in O(n \lg n)$؟ سعیمان را می‌کنیم.

<p></p>
می‌خواهیم به صورت استقرایی نشان دهیم که 
$T(n) \leq c n \log n$
داریم:

$$T(n) \leq 2 T(n/2) + d n \log \log n \leq 2 \times c \times \frac{n}{2} \times (\log n - 1) + d n \log \log n = c n \log n - cn + d n \log \log n$$

و عبارت آخر فقط در صورتی کوچک‌تر از 
$c n \log n$
می‌شود که بتوانیم $c$ را طوری انتخاب کنیم که 
$d n \log \log n - cn$ 
برای هر $n$ کوچک‌تر از صفر شود و می‌دانیم که این کار غیرممکن است. 
(چرا؟)
<p></p>
پس استقرای ما به بن‌بست خورد. اگر 
$T(n) \not \in O(n \lg n)$
هر کار دیگری هم که بکنیم به شکست می‌انجامد اما در این مورد خاص واقعا 
$T(n) \in O(n \lg n)$
ولی استقرای ما قدرت کافی برای اثبات آن را ندارد. شیوه‌ای که در این‌جا برای برآمدن از پس این مشکل استفاده می‌کنیم ممکن است عجیب جلوه کند اما وقتی به روند استقرا دقت کنید ایده‌ی جالبی که پشت آن است را می‌بینید.
<p></p>
کاری که می‌کنیم این است که یک جمله‌ی کم‌ارزش‌تر را که خودمان به شکلی هوشمندانه انتخاب کرده‌ایم از فرض و حکم استقرا کم می‌کنیم. در این مورد خاص به جای این که سعی کنیم با استقرا ثابت کنیم که 
$T(n) \leq c n \log n$
سعی می‌کنیم با استقرا ثابت کنیم که 
$T(n) \leq c n \log n - a n \log \log n$.

در این‌جا $a$ یک عدد ثابت است که مثل $c$ بعد از استقرا انتخابش می‌کنیم.

بدیهی است که حکم دوم اولی را نتیجه می‌دهد و اگر آن را اثبات کنیم نشان داده‌ایم که 
$T(n) \in O(n \log n)$
اما اتفاق جالبی که در عمل می‌افتد این است که اثبات این حکم که به نظر سخت‌تر می‌آید، در واقع آسان‌تر است. به استقرا دقت کنید:

$$T(n) \leq 2 T(\frac{n}{2}) + d n \lg \lg n \leq 2 (c \times \frac{n}{2} \times \lg(\frac{n}{2}) - a \times \frac{n}{2} \times \log \log (\frac{n}{2})) + d n \lg \lg n$$

$$ = c n \log n - cn - a n \log \log \frac{n}{2} + d n \lg \lg n$$

حال کافیست $a$ و $c$ را طوری انتخاب کنیم که برای $n$های به قدر کافی بزرگ داشته باشیم:
$$d n \lg \lg n - a n \log \log \frac{n}{2} - c n < 0$$

در این حالت کافیست که مثلا $a$ را بزرگ‌تر از $2d$ انتخاب کنیم.

</font></div>

<font color=#FF7500>
قضیه‌ی ساندویچ
</font><p></p>
حال یک قضیه مطرح می‌کنیم که به قضیه‌ی ساندویچ معروف است و در حقیقت حالت دو طرفه‌ی همان چیزیست که در مورد فرض‌های سازگار بیان کردیم. قضیه‌ی ساندویچ بیان می‌کند که اگر برای $n$ های به اندازه‌ی کافی بزرگ داشته باشیم
$c_1 f(n) \leq g(n) \leq c_2 h(n)$
و 
$f(n) \in \Omega(i(n))$
و 
$h(n) \in O(i(n))$
آن‌گاه 
$g(n) \in \Theta(i(n))$.
این قضیه را می‌توان به راحتی از خاصیت تعدی 
$O$
و
$\Omega$
به عنوان روابطی تعریف شده بر یک کلاس از تابع‌ها نتیجه گرفت.

<p></p>

به عنوان یک مثال فرض کنید تابع بازگشتی زیر به شما داده شده است:
$$T(n) = 2 T(n/2) + \Theta(n \log \log \log n)$$

حال دو تابع زیر را در نظر بگیرید:

$$U(n) = 2 U(n/2) + \Theta(n)$$
$$V(n) = 2 V(n/2) + \Theta(n \log \log n)$$

با تحلیل‌هایی مشابه مثال‌های قبل می‌توان نشان داد که
$U \in \Omega(n \log n)$ و $V \in O(n \log n)$ و در نتیجه $T \in \Theta(n \log n)$.

<p></p>
حال این مثال را که بعدها کاربردی خواهد بود بررسی می‌کنیم:
$$
T(n) = \left\{\begin{matrix}
1 & n = 1 \\ 
T(n-1) + \frac{1}{n} & n > 1 
\end{matrix}\right.$$

شاید بهتر باشد که این تابع را این‌طور بنویسیم:
$$T(n) = \sum_{i=1}^n \frac{1}{i}$$

آن‌گاه داریم:
$$\int_{1}^{n} \frac{1}{x} dx \leq T(n) \leq \int_{2}^{n+1} \frac{1}{x} dx$$

برای این که این موضوع را ببینید به نمودار تابع 
$y = \frac{1}{x}$
که در این تصویر داده شده دقت کنید. نقاط متناظر با تابع 
$T(n)$
را با ستاره نشان داده‌ایم. به سادگی می‌توان دید که این نقاط یک افراز ریمانی واحد برای هر کدام از انتگرال‌های مطرح‌شده تشکیل می‌دهند که برای یکی یک کران پایین و برای دیگری یک کران بالا به دست می‌دهد.

 در نتیجه طبق قضیه‌ی ساندویچ
$T(n) \in \Theta(\lg n)$.
<p></p>

به عنوان یک مثال دیگر فرض کنید:
$$T(n) = \sum_{i=1}^n \log i$$
می‌خواهیم ثابت کنیم که 
$T(n) \in \Theta(n \lg n)$. داریم:
$$T(n) = \sum_{i=1}^n \lg i > \sum_{i=\left \lceil \frac{n}{2} \right \rceil}^n \lg i > (\frac{n}{2} - 1)(\lg n - 1)$$
از طرفی بدیهیست که 
$T(n)\leq n \lg n$ و لذا 
$T(n) \in \Theta(n \lg n)$.
</font></div>

<font color=#FF7500>
روش تغییر متغیر
</font><p></p>
تکنیک بعدی که به آن می‌پردازیم استفاده از تغییر متغیر است. بسیار پیش می‌آید که حل یک رابطه‌ی بازگشتی با تعریف متغیرهای جدید و نوشتن تابع بر اساس آن‌ها آسان‌تر شود. در این‌جا به دو نمونه می‌پردازیم:

$$T(n) = 2 T(\sqrt n) + \lg n$$

پیدا کردن پیچیدگی این تابع با حدس و سپس اثبات آن چندان ساده نیست اما می‌توانیم از تغییر متغیر استفاده کنیم. فرض کنید که قرار دهیم
$n = 2^m$ یا معادلا $m = \lg n$. (آیا این فرض سازگار است؟ چرا؟) حال عبارت بالا را بر اساس متغیر جدید می‌نویسیم:

$$T(2^m) = 2 T(2^{\frac{m}{2}}) + m$$

حال تابع جدید 
$S$
را به این شکل تعریف می‌کنیم:
$$S(m) = T(2^m)$$

با این تعریف می‌توانیم بنویسیم:
$$S(m) = 2 S(m/2) + m$$

این یک رابطه‌ی بازگشتی بسیار ساده‌تر است که می‌توانیم با تکنیک‌های قبلی حلش کنیم و به دست بیاوریم که 
$S(m) \in O(m \lg m)$ اما 
$S(m) = T(2^m) = T(n)$
پس
$T(n) \in O(m \lg m)$
یا معادلا
$$T(n) \in O(\lg n \lg \lg n)$$
و این تحلیل را تمام می‌کند. دقت کنید که همین روش برای $\Omega$ و در نتیجه $\Theta$ هم جواب می‌دهد.

<p></p>
حال همین مثال را کمی تغییر می‌دهیم و سعی می‌کنیم این تابع را تحلیل کنیم:
$$T(n) = 3 T(\sqrt n) + \lg n$$
اگر همان تغییر متغیر قبلی را اعمال کنیم به این رابطه می‌رسیم:
$$S(m) = 3 S(m/2) + m$$
که گرچه با روش‌هایی که تا حالا بیان شدند قابل تحلیل است اما حدس زدن جوابش چندان ساده نیست. خوش‌بختانه یک راه کلی برای به دست آوردن راه حل توابع بازگشتی از این فرم وجود دارد که یک بار برای همیشه حدس زدن را انجام داده و حدس را ثابت کرده و باعث می‌شود به سادگی بتوانیم جواب این رابطه را بیابیم. این راه کلی را قضیه‌ی اصلی می‌نامیم و بخش بعدی به آن می‌پردازد.
</font></div>

<font color=#FF7500>
قضیه‌ی اصلی
</font>
<p></p>
اگر تابع بازگشتی به صورت زیر داشته باشیم:
$$T(n)=aT(\frac{n}{b})+f(n)$$
آن‌گاه چهار حالت ممکن است رخ دهد:
<ul>
<li>وجود داشته باشد عدد $\epsilon>0$ بطوری که 
$n^{\log^a_b+\epsilon} \in O(f(n))$. یعنی 
$n^{\log^a_b}$ به طور چند جمله‌ای از $f(n)$ کم‌تر باشد. در این صورت خواهیم داشت: $T(n)\in \Theta(f(n))$
</li>
<li>وجود داشته باشد عدد $\epsilon>0$ بطوری که 
$f(n) \in O(n^{\log^a_b+\epsilon})$. یعنی 
$f(n)$ به طور چند جمله‌ای از $n^{\log^a_b}$ کم‌تر باشد. در این صورت خواهیم داشت: $T(n)\in \Theta(n^{log^a_b})$
</li>
<li> $n^{\log^a_b} \in \Theta(f(n))$
یعنی $f(n)$ و $n^{\log^a_b}$ مرتبه‌ی یکسانی داشته باشند. در این صورت داریم 
$T(n)\in \Theta(f(n)\lg(n))=\Theta(n^{\log^a_b} \lg(n))$
</li>
<li>هیچ یک از سه حالت فوق رخ ندهد. در این صورت از قضیه‌ی اصلی نمی‌توان استفاده کرد و برای یافتن پیچیدگی تابع بازگشتی باید سراغ روش‌های دیگر رفت. </li>
</ul>
</font></div>

به عنوان نخستین مثال از تحلیل زمان اجرا به وسیله‌ی قضیه‌ی اصلی، الگوریتم مرتب‌سازی ادغامی را مرور می‌کنیم.

در این الگوریتم برای مرتب‌سازی یک آرایه به طول $n$ ابتدا آن را به دو بخش تقریبا مساوی نصف می‌کنیم. سپس هریک از این دو بخش را به صورت بازگشتی مرتب می‌کنیم. در انتها این دو قسمت را ادغام می‌کنیم. برای این کار ابتدا به کوچک‌ترین عنصر دو لیست نگاه می‌کنیم و کم‌ترین آن‌ها را به عنوان کمترین عنصر کل آرایه در نظر می‌گیریم و آن را از لیستی که در آن قرار داشته حذف می‌کنیم. دوباره به کمترین عنصر باقیمانده در دو لیست نگاه می‌کنیم و دومین عنصر کل آرایه را می‌یابیم. این کار را تا پایان یکی از لیست‌ها انجام می‌دهیم. در نهایت کافی است انتهای لیست دیگر را به انتهای آرایه‌ی مرتب شده اضافه کنیم تا کل عناصر مرتب شوند.
</font></div>

```python
def mergesort(A):
    if len(A) < 2: 
        return A
    mid = len(A) // 2
    B = A[:mid]
    C = A[mid:]
    #print(B, C) # Remove comment to see running of algorithm
    B = mergesort(B)
    C = mergesort(C)
    i = j = 0
    A = []
    while i < len(B) and j < len(C):
        if B[i] <= C[j]:
            A += [B[i]]
            i += 1
        else:
            A += [C[j]]
            j += 1
    A += B[i:] + C[j:]
    #print(A) # Remove comment to see running of algorithm
    return A
```

```python
A=[5, -1, 3, 2, -4, 2, 8, 1, 0, -7, 9, 6, 1, 4]
A = mergesort(A)
print(A)
```

اگر $T(n)$ را زمان اجرای مرتب‌سازی ادغامی برای یک آرایه به طول $n$ در نظر بگیریم در این صورت داریم:
$$T(n)=2T(\frac{n}{2})+\Theta(n)$$
اگر با قضیه‌ی اصلی بخواهیم به این مساله نگاه کنیم خواهیم داشت:
$$n^{\log^a_b}=n^{\log^2_2}=n$$
$$f(n) \in \Theta(n)$$
و در نتیجه:
$$n^{\log^a_b}\in \Theta(f(n))$$
لذا طبق قضیه‌ی اصلی داریم:
$$T(n)\in \Theta(n^{\log^a_b}\lg(n)) = \Theta(n\lg(n))$$

<p></p>
<font color=#FF7500>
مثال:
</font>
تابع زیر را با استفاده از قضیه‌ی اصلی تحلیل کنید:
$$T(n)=7T(n/3)+O(n)$$
در این حالت $n^{\log^a_b-\epsilon}=n^{\log^7_3-\epsilon}>n$
برای همه‌ی مقادیر 
$0 < \epsilon < \log^7_3-1$ 
برقرار است. پس خواهیم داشت:
$$T(n)\in\Theta(n^{\log^a_b})=\Theta(n^{\log^7_3})\approx\Theta(n^{1.77})$$
</font></div>

$$T(n)=8T(\frac{n}{2})+25\times n^3$$
در این حالت $n^{\log^a_b}=n^{\log^8_2}\in \Theta(25\times n^3)$
<p></p>
لذا طبق قضیه‌ی اصلی خواهیم داشت:
$$T(n)\in\Theta(n^{\log^a_b}\lg(n))=\Theta(n^{\log^8_2}\lg(n))=\Theta(n^3\lg(n))$$

<font color=#FF7500>
مثال:
</font>
تابع زیر را با استفاده از قضیه‌ی اصلی تحلیل کنید:
$$T(n)=4T(n/2)+O(n^2\lg(n))$$
با وجود آن که $n^{\log^a_b}=n^{\log^4_2}=n^2\in O(n^2\lg(n))$
ولی هیچ مقدار $\epsilon>0$ وجود ندارد که به ازای آن
$n^{2+\epsilon}\in O(n^2\lg(n))$ 
چرا که هرچقدر هم مقدار کمتری برای $\epsilon$ انتخاب کنیم بازهم مقداری از $n_0$ وجود خواهد داشت که برای $n>n_0$ داشته باشیم  $n^{\epsilon}>\lg(n)$.
به همین دلیل هیچ یک از دو تابع $n^2$ و $n^2\lg(n)$ به طور چند جمله‌ای بیشتر از تابع دیگر نیست، و این دوتابع روند رشد یکسانی هم ندارند، لذا نمی‌توانیم از قضیه‌ی اصلی استفاده کنیم.

بعضی اوقات با رسم یک درخت که روند تابع بازگشتی را در هر سطح نشان دهد می‌توان به سادگی توابع بازگشتی را تحلیل کرد. مثلا فرض کنید بخواهیم تابع زیر را تحلیل کنیم.
$$T(n)=T(\frac{9n}{10})+T(\frac{n}{10})+n$$
$$T(1)=1$$
می‌توانیم روند اجرای این رویه‌ی بازگشتی را به صورت یک درخت در نظر بگیریم که در ریشه‌ی آن مقدار $n$ قرار دارد. هزینه‌ی $T(n)$ به صورت بازگشتی مجموع هزینه‌های $T(\frac{9n}{10})$ و 
$T(\frac{n}{10})$ است، لذا می‌توانیم برای ریشه دو راس با اندازه‌های $\frac{9n}{10}$ و $\frac{n}{10}$ در نظر بگیریم. علاوه بر این هزینه‌های بازگشتی، تابع $T(n)$ یک هزینه‌ی $n$ هم دارد که می‌توانیم آن را به صورت جداگانه در سطح ریشه محاسبه کنیم. با این تفاسیر به شکل زیر می‌رسیم:
